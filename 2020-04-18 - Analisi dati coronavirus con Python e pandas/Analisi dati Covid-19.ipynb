```python
df.sum()
```

```python
df_plot.plot(kind="bar", x="denominazione_regione", figsize=(20,10))
```

```python
df_world = pd.read_excel("COVID-19.xlsx")
```

```python
df_world
```

```python
df_world_sum = df_world.groupby(by="countriesAndTerritories").sum().reset_index()
```

```python
df_world_sum[df_world_sum["countriesAndTerritories"]=="Italy"]
```

```python
df_world_sum[df_world_sum["cases"]>=20000].sort_values(by="cases", ascending=False).plot(kind="bar", x="countriesAndTerritories", y = ["cases","deaths"], figsize=(20, 10))
```
