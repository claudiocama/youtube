```python
from pandas_datareader import data
```

```python
df = data.get_data_yahoo("SPY", start="2010-05-01")
```

```python
df
```

```python
VERSAMENTO = 300
```

```python
df_m = df.asfreq("m", method="bfill")
```

```python
df_m
```

```python
df_m.reset_index(inplace=True)
```

```python
df_m
```

```python
df_m["Cash"] = 0.0
df_m["Acquisti"] = 0
```

```python
df_m
```

```python
df_m.loc[120] = 0
```

```python
df_m
```

```python
df_m = df_m.shift(fill_value=0)
```

```python
df_m
```

```python
for index in range(1,len(df_m)):
    df_m["Acquisti"][index] = (df_m["Cash"][index-1]+VERSAMENTO)//df_m["Adj Close"][index]
    df_m["Cash"][index] = (df_m["Cash"][index-1]+VERSAMENTO)%df_m["Adj Close"][index]
```
