```python
%matplotlib inline
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df = yf.download("AAPL", start="2015-01-01")
```

```python
df
```

```python
df.plot()
```

```python
df.plot(y="Adj Close", figsize=(15, 8))
```

```python
top = plt.subplot2grid((4,1), (0, 0), rowspan=3, colspan=1)
top.plot(df.index, df["Adj Close"])

bottom = plt.subplot2grid((4,1), (3,0), rowspan=1, colspan=1)
bottom.bar(df.index, df['Volume'])

plt.gcf().set_size_inches(15,8)
```
